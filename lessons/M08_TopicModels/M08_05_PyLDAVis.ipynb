# Setup

## Config

```python
import configparser
config = configparser.ConfigParser()
config.read("../../../env.ini")
output_dir = config['DEFAULT']['output_dir']
data_prefix = 'austen-melville'
```

## Import

```python
import pandas as pd
import numpy as np
import pyLDAvis
import pyLDAvis.lda_model
import pickle
```

# Import Data

```python
lda_engine = pickle.load(open(f'{output_dir}/{data_prefix}-lda_engine.pickle', 'rb'))
count_engine = pickle.load(open(f'{output_dir}/{data_prefix}-count_engine.pickle', 'rb'))
count_model = pickle.load(open(f'{output_dir}/{data_prefix}-count_model.pickle', 'rb'))
```

# Visualize

```python
pyLDAvis.enable_notebook()
```

```python
pyLDAvis.lda_model.prepare(lda_engine, count_model, count_engine)
```
